# Exercise 18.1

Analyze the writing patterns of each student

* Puntuación 
* Cantidad de palabras
* Cantidad de oraciones

```python
bd = data.copy()
```

```python
bd.replace(np.nan, "Faltante", inplace=True)
bd.head()
```

#### Variable conteo de palabras

```python
bd['palabras'] = bd['T1'].str.split().str.len()+bd['T2'].str.split().str.len()+bd['T3'].str.split().str.len()+bd['T4'].str.split().str.len()+bd['T5'].str.split().str.len()+bd['T6'].str.split().str.len()
bd.head()
```

#### Variable conteo de tareas faltantes

```python
bd["faltantes"]= bd['T1'].str.count("Faltante")+bd['T2'].str.count("Faltante")+bd['T3'].str.count("Faltante")+bd['T4'].str.count("Faltante")+bd['T5'].str.count("Faltante")+bd['T6'].str.count("Faltante")
bd.head()
```

#### Promedio de palabras por texto

```python
bd['promedio_palabras']= bd['palabras']/(6-bd['faltantes'])
bd.head(3)
```

#### Promedio de puntuación por texto

```python
import string
string.punctuation
```

```python
bd['puntuacion_T1'] = (bd['T1'].str.count("\.")+bd['T1'].str.count("\,")+bd['T1'].str.count("\?")+bd['T1'].str.count('\"')+bd['T1'].str.count('\¿'))
bd['puntuacion_T2'] = (bd['T2'].str.count("\.")+bd['T2'].str.count("\,")+bd['T2'].str.count("\?")+bd['T2'].str.count('\"')+bd['T2'].str.count('\¿'))
bd['puntuacion_T3'] = (bd['T3'].str.count("\.")+bd['T3'].str.count("\,")+bd['T3'].str.count("\?")+bd['T3'].str.count('\"')+bd['T3'].str.count('\¿'))
bd['puntuacion_T4'] = (bd['T4'].str.count("\.")+bd['T4'].str.count("\,")+bd['T4'].str.count("\?")+bd['T4'].str.count('\"')+bd['T4'].str.count('\¿'))
bd['puntuacion_T5'] = (bd['T5'].str.count("\.")+bd['T5'].str.count("\,")+bd['T5'].str.count("\?")+bd['T5'].str.count('\"')+bd['T5'].str.count('\¿'))
bd['puntuacion_T6'] = (bd['T6'].str.count("\.")+bd['T6'].str.count("\,")+bd['T6'].str.count("\?")+bd['T6'].str.count('\"')+bd['T6'].str.count('\¿'))
bd.head()
```

```python
bd['prom_puntuacion'] = (bd['puntuacion_T1']+bd['puntuacion_T2']+bd['puntuacion_T3']+bd['puntuacion_T4']+bd['puntuacion_T5']+bd['puntuacion_T6'])/(6-bd['faltantes'])
bd.head()
```

#### Cantidad de oraciones

```python
bd['oraciones_T1'] = bd['T1'].str.count("\.")
bd['oraciones_T2'] = bd['T2'].str.count("\.")
bd['oraciones_T3'] = bd['T3'].str.count("\.")
bd['oraciones_T4'] = bd['T4'].str.count("\.")
bd['oraciones_T5'] = bd['T5'].str.count("\.")
bd['oraciones_T6'] = bd['T6'].str.count("\.")
bd['total_oraciones'] = bd['oraciones_T1']+bd['oraciones_T2']+bd['oraciones_T3']+bd['oraciones_T4']+bd['oraciones_T5']+bd['oraciones_T6']
bd['prom_oraciones'] = bd['total_oraciones']/(6-bd['faltantes'])
bd.head()
```

# Exercise 18.3

Create a classifier to predict the sex of each student



```python
# define X and y
feature_cols = ['palabras', 'faltantes', 'promedio_palabras', 'prom_puntuacion', 'prom_oraciones']
X = bd[feature_cols]
y = bd.Sexo
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# fit a classification tree with max_depth=3 on all data
from sklearn.tree import DecisionTreeClassifier
treeclf = DecisionTreeClassifier(max_depth=None, random_state=1)
treeclf.fit(X_train, y_train)
```

```python
# compute the feature importances
pd.DataFrame({'feature':feature_cols, 'importance':treeclf.feature_importances_})
```

```python
y_pred = treeclf.predict(X_test)
from sklearn.metrics import f1_score
f1_score(y_test, y_pred, average='micro')  
```
